** 1- what is the rate of reservtion per month and year?**

```python
plt.figure(figsize=(12,4))
sns.countplot(x= 'arrival_date_month', palette= "deep", hue = 'hotel' , data = df)
plt.title('Months of Arrival')
plt.show()
```

```python
plt.figure(figsize=(12,4))
sns.countplot(x= 'arrival_date_year', palette= 'crest', hue = 'hotel' , data = df)
plt.title('year of Arrival')
plt.show()
```

**2-countries with most vistors**

```python
country_data = pd.DataFrame(df[df.is_canceled == 0]['country'].value_counts())
country_data.rename(columns= {'country': 'num of guests' }, inplace = True)
country_data['guests percentage'] = round(country_data['num of guests'] / country_data['num of guests'].sum()*100)
country_data['country'] = country_data.index

country_data = country_data.head(10)

fig = px.pie(country_data, values='num of guests', names= 'country')
fig.update_layout(title_text='Countries prefered for Hotel Booking')
fig.update_traces(textposition='inside',textinfo='label+percent+value')
fig.show()
```

3- canceletion rate per hotel

```python
plt.figure(figsize = (10, 5))
sns.countplot(x= 'hotel',data = df, hue = 'is_canceled', palette = 'Blues', alpha = 0.8).set(xlabel = None)
plt.title('Cancelation rates by hotel')
plt.show()
```

4- perfered meal

```python
meal_count = pd.DataFrame(df['meal'].value_counts())
meal_count.rename(columns= {'meal': 'count' }, inplace = True)
meal_count['meal %'] = round(meal_count['count'] / meal_count['count'].sum()*100)
meal_count['meal'] =meal_count.index

fig = px.pie(meal_count, values='count', names= 'meal', hole=0.5,color_discrete_sequence=px.colors.sequential.Jet)
fig.update_layout(title_text='% in Meal', title_x=0.5,height = 700)
fig.update_traces(textposition='inside',textinfo='label+percent+value')
fig.show()
```
